### c)

Let $f(x) = x + x^2/2$, we can see that $x$ is odd and $x^2/2$ is even and can find their corresponing sine and cosine series.
$$
\begin{split}
b_k = \int_{-1/2}^{1/2} x\sin (2\pi kx) dx
&= \left[ -\frac{x}{2\pi k} \cos(2\pi kx) \right]_{-1/2}^{1/2} + \frac{1}{2\pi k} \int_{-1/2}^{1/2} \cos(2\pi kx)dx \\
&= -\frac{1}{4\pi k} \left[ \cos(\pi k) + \cos(-\pi k) \right] + \frac{1}{2\pi k} \left[ \frac{\sin(2\pi kx)}{2\pi k} \right]_{-1/2}^{1/2} \\
&= -\frac{(-1)^k}{2\pi k} + \frac{1}{4\pi^2 k^2} \left[ \sin(\pi k) - \sin(-\pi k) \right] \\
&= -\frac{(-1)^k}{2\pi k}
\end{split}
$$
and
$$
\begin{split}
a_k = \int_{-1/2}^{1/2} \frac{x^2}{2} \cos(2\pi kx) dx
&= \left[ \frac{x^2}{4\pi k} \sin(2\pi kx) \right]_{-1/2}^{1/2} - \frac{1}{2\pi k} \int_{-1/2}^{1/2} x\sin(2\pi kx)dx \\

\end{split}
$$
lastly
$$
a_0 = \int_{-1/2}^{1/2} \left( x + \frac{x^2}{2} \right) dx = \left[ \frac{x^2}{2} + \frac{x^3}{6} \right]_{-1/2}^{1/2} = \frac{2}{8} + \frac{2}{48} = \frac{7}{24}
$$
and we have the Fourier series
$$
f(x) = \frac{7}{24} + \sum_{k=1}^\infty \frac{(-1)^k}{2\pi k} \sin(2\pi kx) + \sum_{k=1}^\infty \frac{(-1)^k}{4\pi^2 k^2} \cos(2\pi kx).
$$
We now find the exponential form by
$$
\begin{split}
c_k &= \int_{-1/2}^{1/2} \left( x + \frac{x^2}{2} \right) e^{2\pi ikx} dx \\
&= \left[ \left( x + \frac{x^2}{2} \right) \frac{1}{2\pi ik} e^{2\pi ikx} \right]_{-1/2}^{1/2} - \frac{1}{2\pi ik} \int_{-1/2}^{1/2} (1 + x) e^{2\pi ikx} dx \\
&= \frac{5}{16\pi ik} e^{\pi ik} - \frac{3}{16\pi ik}e^{-\pi ik} - \frac{1}{2\pi ik} \left( \left[ \frac{1+x}{2\pi ik} e^{2\pi ikx} \right]_{-1/2}^{1/2} - \int_{-1/2}^{1/2} e^{2\pi ikx} dx \right) \\
&= \frac{5}{16\pi ik} e^{\pi ik} - \frac{3}{16 \pi ik} e^{-\pi ik} + \frac{3}{8 \pi^2 k^2} e^{\pi ik} + \frac{1}{8 \pi^2 k^2} e^{-\pi ik} + \frac{1}{2\pi ik} \left[ \frac{1}{2\pi ik} e^{2\pi ikx} \right]_{-1/2}^{1/2} \\
&= \frac{5}{16\pi ik} e^{\pi ik} - \frac{3}{16 \pi ik} e^{-\pi ik} + \frac{3}{8 \pi^2 k^2} e^{\pi ik} + \frac{1}{8 \pi^2 k^2} e^{-\pi ik} - \frac{1}{4\pi^2 k^2} e^{\pi ik} - \frac{1}{4\pi^2 k^2} e^{-\pi ik} \\
&= \frac{(-1)^k}{8\pi ik}
\end{split}
$$
as $e^{\pi ik} = e^{-\pi ik} = (-1)^k, \; \forall k \in \mathbb{Z}$. We then have the exponential Fourier series
$$
f(x) = \sum_{k\in \mathbb{Z}} \frac{(-1)^k}{8\pi ik} e^{2\pi ikx}
$$

$$

$$

thus
$$
\mathbf{u}_2 = x-1.

```python
import matplotlib.pyplot as plt


def chebyshev(k, n=100):
    x = np.linspace(-1, 1, n)
    return np.cos(k*np.arccos(x))


x = np.linspace(-1, 1, 100)

for k in range(7):
    plt.plot(x, chebyshev(k, 100), label=f"{k=}")

plt.legend()
plt.show()


def chebyshev_rec(n, x):
    if n == 0:
        return np.ones(len(x))
    elif n == 1:
        return x
    return 2*x*chebyshev_rec(n-1, x) - chebyshev_rec(n-2, x)


for n in range(7):
    plt.plot(x, chebyshev_rec(n, x), label=f"{n=}")

plt.legend()
plt.show()
```
