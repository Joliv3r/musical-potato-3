```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.array([1+1j, -1+2j, 4-3j])
# x = n.real
# y = n.imag

# xi = (n*1j).real
# yi = (n*1j).imag

for i, j, color in zip(n, n*1j, ['r', 'b', 'y']):
    plt.plot(i.real, i.imag, 'o'+color) 
    plt.plot(j.real, j.imag, 'o'+color)

# plt.plot(x, y, 'o')
# plt.plot(xi, yi, 'o')
plt.plot([0, 0], [-10, 10], 'k')
plt.plot([-10, 10], [0, 0], 'k')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.savefig('oppg1.pdf')
plt.show()
```
