```python
import numpy as np
import matplotlib.pyplot as plt

def a(n):
    if n == 0:
        return 2/np.pi
    elif n == 1:
        return 0
    else:
        return (2*((-1)**(n+1) - 1))/(np.pi*(n**2 - 1))

def b(n):
    if n == 0 or n == 1:
        return 0
    return 0


def F(x, n):
    s = 0
    for i in range(0, n+1):
        s += a(i) * np.cos(i*x) + b(i) * np.sin(i*x)
    return s


x = np.linspace(-10, 10, 100)

plt.plot(x, F(x, 1000))
plt.plot(x, np.sin(x))
plt.show()
```
