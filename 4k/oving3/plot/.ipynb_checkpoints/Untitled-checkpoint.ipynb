```python
import numpy as np
import matplotlib.pyplot as plt

def a(n):
    if n == 0:
        return 4*np.pi**2 / 3
    return 4/(n**2)

def b(n):
    if n == 0:
        return 0
    return -4*np.pi/n


def F(x, n):
    s = 0
    for i in range(0, n+1):
        s += a(i) * np.cos(i*x) + b(i) * np.sin(i*x)
    return s


x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, F(x, 5))
plt.plot(x, x**2)
plt.show()
```
